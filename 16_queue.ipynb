```python
class Queue:
    def __init__(self, size = 5):
        self.queue = [] 
        self.size = size
        self.rear = 0  
        self.front = 0 
        
    def add(self, data):
        #This queue doesn't allow repetion of the same data
        if data in self.queue:
            print("Data {} already exists in the queue".format(data))
        else:
            if self.size > self.rear:
                #overflow가 발생하지 않았다
                self.queue.append(data)
                self.rear += 1
                print("Saving {} in the queue".format(data), end=' ')
                print('rear = {}, front = {}'.format(self.rear, self.front))
            else:
                print("overflow!! Unable to save") 
        self.view()
            
    
    def remove(self):
        if self.rear <= 0 or self.rear == self.front:
            print("Underflow!!")
        else:
            data = self.queue[self.front]
            self.queue[self.front] = ''
            self.front += 1
            print('Data {} removed from the queue'.format(data),end= ' ')
            print("rear = {} front = {}".format(self.rear, self.front))
    def view(self):
        print("Data in the queue => ", end='')
        if self.rear <= 0 or self.rear == self.front:
            print("No data in the queue")
        else:
            for i in range(self.front, self.rear):
                print(self.queue[i], end='   ')
        print()
    
queue = Queue()
queue.add(111)
queue.add(111)
queue.add(112)
queue.add(113)
queue.add(114)
queue.add(115)
queue.add(116)
queue.remove()
queue.remove()
queue.remove()
queue.view()
queue.remove()
queue.view()
queue.remove()
queue.remove()



```
