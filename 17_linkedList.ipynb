```python
#linkedList
#Node class (saving data in nodes and then organizing the nodes in linkedlist)
class Node:
    def __init__(self, data = None):
        self.data = data    #linkedList에 저장할 실제 데이터
        self.next = None   #다음에 처리할 데이터의 위치 (주소)
```

```python
#linkedList class
class LinkedList:
    def __init__(self):
        self.count =  0      
        self.head = None    

    def appendLast(self, data):
        newNode = Node(data)
        self.count += 1
        
        if self.head is None: 
            self.head = newNode
            return
        
        start = self.head
        
        while start.next:     
            start = start.next 
            
        start.next = newNode
        
    def insertFirst(self, data):
        newNode = Node(data)
        self.count += 1
        newNode.next = self.head
        self.head = newNode
    
    
    def insertPosition(self, position, data):
        if position < 1 or position > self.count - 1:
            print('Cannot save data {0} in position {1}'.format(data, position))
            return
        
        newNode = Node(data)
        self.count += 1
        
        start = self.head
        for i in range(position - 1):
            start = start.next
        newNode.next = start.next
        start.next = newNode
            
    def listPrint(self):
        start = self.head
        if start is None:
            print("No saved data in the linkedList")
            
        else:
            print("There are {} data saved in the linkedList:".format(self.count), end = ' ')
            for i in range(self.count):
                print(start.data, end=' ')
                start = start.next
            print()
            
    def remove(self, data):
        start = self.head
        if start == None:
            print("linkedListis empty")
            
        else:
            
            if start.data == data:
                self.head = start.next
                self.count -= 1
                return
            
        while start is not None:
            
            if start.data == data:
                break
            prev = start         
            start = start.next   
        
        if start == None:
            print('Cannot remove data because that data does not exist in the linkedList')
            return
        
        prev.next = start.next
        self.count -= 1
            
```

```python
print('Creating a new linkedList.')
linkedList = LinkedList()
linkedList.listPrint()
linkedList.remove('Cat')
print('=' * 80)

linkedList.appendLast('Cat')
linkedList.listPrint()
print('=' * 80)

linkedList.appendLast('Dog')
linkedList.appendLast('Elephant')
linkedList.appendLast('Tiger')
linkedList.listPrint()
print('=' * 80)

linkedList.insertFirst('Lion')
linkedList.listPrint()
print('=' * 80)

linkedList.insertPosition(0,'Ant')
linkedList.insertPosition(5,'Bee')
linkedList.insertPosition(4,'Crocodile')
linkedList.listPrint()
print('=' * 80)

linkedList.remove('Bee')
linkedList.listPrint()
print('=' * 80)

linkedList.remove("Ant")
linkedList.listPrint()
print('=' * 80)

linkedList.remove('Parrot')
print('=' * 80)
```
